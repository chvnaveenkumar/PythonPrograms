### Use Cases:



```python
### pip install -r .\requirements.txt
from pydantic import BaseModel
```

```python
from dataclasses import dataclass
@dataclass
class Person():
    name: str
    age: int
    city: str

person = Person(name="Naveen", age= 25, city="Nashville")
print(person)

person1 = Person(name="Naveen", age= 25, city=15)
print(person1)
```

```python
class Person(BaseModel):
    name: str
    age: int
    city: str

person = Person(name="Naveen", age= 25, city="Nashville")
print(person)

## Why we are inheriting base model
person1 = Person(name="Naveen", age= 25, city=15)
print(person1)
```

### 2. Model with optional fields


```python
from typing import Optional
class Employee(BaseModel):
    id: int
    name: str
    department: str
    salary: Optional[float] = None
    is_active: Optional[bool] = True


emp = Employee(id=123, name="Naveen",department="Engineering")
print(emp)
```
