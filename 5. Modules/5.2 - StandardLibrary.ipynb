```python
## Data Serialization
import json
data={'name':'Naveen', 'age':25}
json_str = json.dumps(data)
print(json_str)
print(type(json_str))

parsed_data=json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
## csv

import csv

with open('example.csv',mode='w', newline='') as file: ## write w+ -> Write append mode
    writer = csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['Naveen',25])


with open('example.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)

```
