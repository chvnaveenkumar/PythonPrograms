```python
## Variables name case sensitive

firstName = "Naveen"
FirstName = "Nav"

print(firstName)
print(FirstName)
firstName==FirstName
```

```python
## Python is dynamically typed, type of a variable is determined at runtime
age = 25
print(age, type(age))
age = "Naveen"
print(age, type(age))
age = 2.5
print(age, type(age))
```

```python
##Type conversion int to String
age = 25
print(type(age))
age_str = str(age)
print(type(age_str))


##Type conversion String to int
age = '25'
print(int(age))


name = 'Naveen'
print(int(name)) ## we cannot convert string to int


```

```python
## Type conversion int to float
height = 5.1
print(int(height))
## Type conversion float to int
height = 5
print(float(height))
```

```python
## User need to enter value
age=input("What is the age?")
print(age, type(age))


age=int(input("What is the age?")) ## It will take only int type
print(age, type(age))
```
