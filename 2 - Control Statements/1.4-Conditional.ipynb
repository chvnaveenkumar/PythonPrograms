```python
# if statement
age = 15
if age>=20:
    print("Age is greater then 20")
else:
    print("Age is less than 20")
```

```python
# if statement
age = 20
if age>=20:
    print("Age is greater then 20")
elif age==20:
    print("Age is equal to 20")
else:
    print("Age is less than 20")
```
