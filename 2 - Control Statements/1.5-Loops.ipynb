```python
## While loop
count =0
while count<5:
    print(count)
    count = count+1
```

```python
count = 0
while count%2==0:
    print(count)
    count = count + 1
```

```python
## Loop control statments
## break
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## Loop control statments
## break
for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
## Pass statement is null operation; it does nothing.

for i in range(5):
    if i == 3:
        print(" The number is ", i)
        pass
    print(i)
```
