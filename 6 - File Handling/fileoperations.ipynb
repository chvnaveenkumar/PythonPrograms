```python
with open('example.txt', mode='r') as file:
    content = file.read()
    print(content)

## We will get error if no file
```

```python
## Read file line by line
with open('example.txt', mode='r') as file:
    for line in file:
        print(line.strip())
```

```python
## Writing a file(overwritting)

with open('example.txt', mode='w') as file:
    file.write('Hello World\n')
    file.write('This is a new line')
    
```

```python
## Writing a file(without overwritting)

with open('example.txt', mode='a') as file:
    file.write('Welcome\n')
    file.write('Example text\n\n')
    
```

```python
## Writing list of line to the file
lines = ['First Line \n','Second Line \n', 'Third Line\n']
with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
### Binary Files


## Writing list of line to the file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
### Read the content from the source text file and write to a destination text file

with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt','w') as destination_file:
    destination_file.write(content)
```

```python
### Writing and then reading 

with open('example.txt','w+') as file: ## w+ with both read and write. If file not exits then it will create a file
    file.write("Hello World\n")
    file.write("This is a new line\n")

    ## Move the file cursor to the beginning (To Test comment file.seek(0))
    file.seek(0)

    content = file.read()
    print(content)
```

```python
import os

path='example.txt'
if os.path.isfile(path):
    print(f"The path '{path}' is file")
elif os.path.isdir(path):
    print(f"The path '{path}' is dierectory")
else:
    print(f"The path '{path}' is not a file or directory")
```
