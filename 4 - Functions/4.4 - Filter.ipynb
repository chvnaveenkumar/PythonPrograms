```python
## filter function filter out items based on a condition

def even(num):
    if(num % 2 == 0):
        return True
    
even(6)
```

```python
lst = [1,2,3,4,5,6,7,8]
list(filter(even, lst))
```

```python
numbers = [1,2,3,4,5,6,7,8]
greater_than_five = list(filter(lambda x:x>5, numbers))
print(greater_than_five)
```

```python
## Filter with lambda fucntion with multiple conditions

numbers = [1,2,3,4,5,6,7,8]
even_and_greater_then_5 = list(filter(lambda x:x>5 and x%2 ==0, numbers))
print(even_and_greater_then_5)
```

```python
## Filter() to check if age is greater then 25 in dictionary

def get_name(person):
    return person['age'] > 25

person=[
    {'name': 'Krish','age':32},
    {'name': 'Jack','age': 45},
    {'name': 'Naveen','age': 20}
]

list(filter(get_name, person))
```
