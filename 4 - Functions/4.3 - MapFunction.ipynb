```python
def square(x):
    return x * x

square(4)
```

```python
numbers=[1,2,3,4,5,6,7,8]
map(square, numbers) ## Two parameters function name and iterable 
print(list(map(square, numbers)))
```

```python
## Lambda function with map

numbers = [1,2,3,4,5,6,7,8]
list(map(lambda x:x*x, numbers))
```

```python
## Map Multiple Iterables

numbers1=[1,2,3]
numbers2=[4,5,6]


added_numbers=map(lambda x,y:x+y, numbers1, numbers2)
print(list(added_numbers))
```

```python
## map() to convert list of strings to integers

str_numbers = ['1','2','3','4','5']
int_numbers = list(map(int, str_numbers))
print(int_numbers)
```

```python
words = ['banana','apple','orange']
upper_words = list(map(str.upper, words))
print(upper_words)
```
