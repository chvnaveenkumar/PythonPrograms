```python
def addition(a,b):
    return a+b

addition(2,5)
```

```python
addition = lambda a,b:a+b
print(addition(2,3))
print(type(addition))
```

```python
addition = lambda x,y,z: x+y+z
print(addition(2,3,4))
```

```python
## map() - It applies function to all items in a list

numbers = [1,2,3,4,5]
squared_numbers = map(lambda x: x**2, numbers)
print(squared_numbers) ## It create object at particular location
print(list(squared_numbers))
```
