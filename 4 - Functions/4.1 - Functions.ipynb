```python
def evenOrOdd(num):
    if num%2 == 0:
        print("Number is even")
    else:
        print("Number is odd")

evenOrOdd(5)
evenOrOdd(6)
```

```python
## Function with multiple Parameters

def add(a,b):
    return a+b

print(add(2,6))
```

```python
## Default Parameters

def greet(name="Guest"):  ## it will check parameter has value or not. if not then take Guest as name
    print(f"Hello: {name}")

greet("Naveen")
greet()
```

```python
## Variable length Arguments
## Positoinal And Keywords arguments

def print_number(*args): ## it is not requried to give args it can be any name
    for numbers in args:
        print(numbers)


print_number(1,2)
print_number(2,3,4,5)
print_number(1,"Hello","Test",301)
```

```python
## Keywords Arguments

def print_details(**kwargs): ## it is not requried to give args it can be any name
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="Krish", age=32, country="India")
```

```python
def print_details(*args, **kwargs): ## it is not requried to give args it can be any name    
    for numbers in args:
        print(f"Positional argument {numbers}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(1,2,3,5,name="Krish", age=32, country="India")
```

```python
## Return statements
def multiply(a,b):
    return a*b

multiply(2,3)
```

```python
## Return multiple values
def multiply(a,b):
    return a*b,a

multiply(2,3)
```
