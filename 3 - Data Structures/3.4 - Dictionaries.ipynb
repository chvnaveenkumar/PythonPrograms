```python
## Modifiying Dictionary Elements
## Dictionary are mutable, so you can add, update or delete elements

student={"name":"Naveen","age":32,"grade":"A"}
print(student)
student["address"]="Nashville"
print(student["age"])
print(student["address"])
print(student)
```

```python
print(student)
del student["grade"] ## delete key and value pair
print(student)
```

```python
## Dictionary Methods

keys = student.keys()
print(keys)
values = student.values()
print(values)

items = student.items()
print(items)
```

```python
## Shallow copy

student_copy = student
print(student_copy)
student_copy["name"]="Kumar"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy() ## It will create new memory
student_copy1["name"] = "Ch"
print(student)
print(student_copy1)
```

```python
## Iterating over Dictionaries
## You can be loops to iterative over dictionaries, keys, values or items

## Iterating over keys

for keys in student.keys():
    print(keys)
```

```python
## Iterate over values
for values in student.values():
    print(values)
```

```python
for key,value in student.items():
    print(f"{key} - {value}")
```

```python
## Nested dictionaries

students={
   "student1":{"name":"Naveen","age":32,"grade":"A"},
   "student2":{"name":"Kumar","age":35,"grade":"B"},
}

print(students)

## Access dictionary elements
print(students["student2"]["name"])
print(students["student2"]["age"])

## Iterating over dictionaries

for student_id, student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key, value in student_info.items():
        print(f"{key}: {value}")
```

```python
## Dictionary Comphrehension

squares = {x:x**2 for x in range(5)}
print(squares)
```

```python
## Conditional dictionary comprehension
evens = {x:x**2 for x in range(10) if x%2==0 }

print(evens)

```

```python
## Use a dictionary to count the frequency of elements in list


numbers =[0,1,1,2,2,3,3,3,4,4,5]

frequency={}

for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1

print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merge_dict={**dict1,**dict2}
print(merge_dict)
```
