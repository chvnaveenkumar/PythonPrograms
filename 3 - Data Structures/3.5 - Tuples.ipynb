```python
## Tuples are ordered collection of items that are immutable. 
## They are similar to the lists, but their immutability makes them different
## creating a tuple
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))

tpl = tuple()
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,5,6])
print(numbers)
print(type(numbers))

mixedTuples=(1,"Hello Word",2.5)
print(mixedTuples)
```

```python
## Accessing tupels
print(numbers[2])
print(numbers[-1])

## Slicing operation also same as list
print(numbers[:4])
print(numbers[::1])
print(numbers[::2])
```

```python
## Tuple Operations
numbers + mixedTuples
```

```python
## Operations It will repeat multiple times
mixedTuples * 2
```

```python
numbers * 3
```

```python
## Immutable nature of Tuples
## Tuples are immutable. Once assigned that cannot be changed

lst2=[1,2,3,4,5]
print(lst2)

lst2[1] = "Naveen"
print(lst2)
```

```python
numbers[1]="Naveen" ## Tuples are immutable. It cannot be changed once assinged
```

```python
## Tuple Methods
print(numbers)
print(numbers.count(1))
print(numbers.index(3)) ## It returns first index of the value
```

```python
## Packing and Unpacking Tuple

pack_tuple =(1,"Hellow",3.12)
print(pack_tuple)

## How do we unpack a tuple
a,b,c = 1, "Hellow", 3.12
print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers =(1,2,3,4,5,6)
first,*middle, last = numbers
print(first)
print(last)
print(middle)
```

```python
## Nested Tuple
## Nested List
lst=[[1,2],[3,4,5],[1,"Hello",3.14,"C"]]
print(lst)
print(lst[0])
```
