```python
## Modifying the list elements
fruits = ["Banana","Orange","Grapes","Cherry","Kiwi","Gauva"]
print(fruits[1:])
fruits.append("Apple")
print(fruits)
```

```python

fruits.insert(1,"Peers")
print(fruits)

fruits.remove("Peers") ## Remove first occurance of the item
print(fruits)

fruits.remove("Apple") ## Remove first occurance of the item
print(fruits)
```

```python
## Remove and retrun the last element
popped_element = fruits.pop()
print(popped_element)
print(fruits)
```

```python
print(fruits.index("Orange"))
fruits.insert(2, "Orange")
print(fruits.count("Orange"))
```

```python
print(fruits)
fruits.sort()
print(fruits)
fruits.reverse()
print(fruits)
```
