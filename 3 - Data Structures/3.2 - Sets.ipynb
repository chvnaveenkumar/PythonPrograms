```python
duplicateElements = {1,2,2,4,3,3,4,5,6,7,8}
print(duplicateElements)
```

```python
## Adding and Removing Elements
duplicateElements.add(9)
print(duplicateElements)
duplicateElements.add(15)
print(duplicateElements)
duplicateElements.remove(3)
print(duplicateElements)
```

```python
duplicateElements.remove(20)
print(duplicateElements)
```

```python
## use discard that doesnt through error and it will remove if exits
duplicateElements.discard(20)
print(duplicateElements)
```

```python
removed_element=duplicateElements.pop()
print(removed_element)
print(duplicateElements)
```

```python
## Clear all the elements
duplicateElements.clear()
print(duplicateElements)
```

```python
## Set Membership Test
## Whether element is member of the set or not
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical Operations
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

### Union
union_set = set1.union(set2)
print(union_set)

### Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

### Intersection Update
intersectionUnion_set = set1.intersection_update(set2)
print(intersectionUnion_set)
```

```python
## Mathematical Operations
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## Difference
print(set1.difference(set2))
print(set2.difference(set1))
```

```python
## Symmentric Difference
set1.symmetric_difference(set2)
```

```python
## Sets Methods
set1={1,2,3}
set2={3,4,5}

## isSubset (All the elements)
print(set1.issubset(set2))
print(set1.issuperset(set2))
```
